# Image Classification

When we talk about image classification, we are essentially referring to the task of training a machine learning model to assign labels or categories to images based on their visual content. This is an important problem in the field of computer vision and has numerous applications in various domains.

## Understanding Images as Pixels

Images are composed of pixels, which are the smallest units of information in a digital image. Each pixel represents a specific color or intensity value. In the case of grayscale images, each pixel typically has a value ranging from 0 to 255, where 0 represents black and 255 represents white.

Color images are represented using multiple channels, typically Red, Green, and Blue (RGB). In this case, each pixel is represented by three values, indicating the intensity of red, green, and blue channels. The combination of these channels creates different colors and shades.

Computers can only understand numerical data, so in order to perform image classification, we need to convert these images into a numerical representation that can be processed by machine learning algorithms.

## Intuition Behind Image Classification

Image classification involves assigning a label or class to an image based on its visual content. Traditional algorithms like logistic regression or k-nearest neighbors treat images as flat vectors of pixel values and overlook the spatial relationships between neighboring pixels. To overcome this limitation, Convolutional Neural Networks (CNNs) were developed.

## What is CNN and Why Use It?

Lets first understand what does convolution means. Convolution is a process which helps to extract features of images, now these features can be edges ,shapes ,size ,colours etc.

CNNs, also known as ConvNets, are a category of Neural Networks that excel in image recognition and classification tasks. They are specifically designed to process grid-like data, such as images. CNNs leverage the spatial relationships between pixels by applying convolutional filters, which extract relevant features from the input image. These filters capture patterns at different scales, enabling the network to learn hierarchical representations of the image. Additionally, CNNs use pooling layers to downsample the feature maps, reducing spatial dimensions and extracting the most important information. Finally, fully connected layers and softmax activation are employed for classification.

## Spatial Relationships Between Pixels

To understand spatial relationships between pixels, imagine solving a picture puzzle with various puzzle pieces, where each piece represents a pixel in an image. Traditional algorithms like logistic regression or k-nearest neighbors treat the puzzle pieces as separate entities and disregard their spatial arrangement. They focus solely on the characteristics of each piece without considering the overall picture.

In contrast, CNNs recognize that the arrangement and relationships between puzzle pieces are crucial for solving the puzzle and recognizing the image. CNNs take into account the spatial layout of the puzzle pieces and consider how they fit together to form the complete picture. By analyzing both individual puzzle pieces (pixels) and the patterns formed by neighboring pieces, CNNs can detect important features such as edges, textures, and shapes. For example, they learn that specific combinations of puzzle pieces, like a curved line or the outline of an ear, indicate the presence of a cat.

## Intuition Behind CNN

To better understand how CNNs work, let's compare them to how humans classify images. Suppose we want to classify an image of a dog as a Golden Retriever. As humans, we identify certain parts of the dog, such as the nose, eyes, and fur. We break down the image into smaller pieces, recognize those pieces, and then combine them to get an idea of the overall dog. For instance, we might identify a nose, two eyes, and golden fur.

Going further, we can break down the concept of a nose. A Golden Retriever's nose can be seen as an oval shape with two black holes inside it. Classifying the nose involves identifying smaller pieces, such as black holes (nostrils) and curves that define an oval shape.

Broadly speaking, this is what a CNN learns to do. It recognizes basic lines and curves, then shapes and blobs, and progressively more complex objects within the image. Finally, the CNN classifies the image by combining larger, more complex objects.

In CNNs, this learning process occurs through forward propagation and backpropagation, without explicitly programming the network to recognize specific features. The CNN learns to identify objects and patterns on its own, resulting in remarkable image classification capabilities.




## Filters/Kernels and Convolution

In CNNs, filters (also known as kernels) are small matrices used to extract features from images. A filter is a matrix of weights that is convolved (slid) over the input image. The convolution operation involves element-wise multiplication of the filter with the corresponding pixels of the input image and summing the results. This process helps extract specific features such as edges, textures, or colors from the image.

For example, consider a 5x5 matrix as an input image and a 3x3 matrix as a filter. The convolution operation involves multiplying each element of the filter with the corresponding element in the input image and summing the results. This produces a single value, which forms a feature map or output feature.

The filter is then moved (with a defined stride) to slide over the entire image, performing the convolution operation at each position. This process helps capture different features by detecting patterns in different parts of the image.

## Pooling

Pooling is a downsampling operation used in CNNs to reduce the spatial dimensions of the feature maps. It helps extract the most salient information while reducing the computational complexity of the network.

![Pooling](http://ufldl.stanford.edu/tutorial/images/Pooling_schematic.gif)
http://ufldl.stanford.edu/tutorial/supervised/Pooling/

A common pooling technique is max pooling, which divides the feature map into small regions (e.g., 2x2) and outputs the maximum value within each region. By selecting the maximum value, the pooling operation retains the most significant feature present in that region.

Pooling helps to make the network more robust to variations in the input image, improves computational efficiency, and introduces a degree of spatial invariance. It reduces the number

 of parameters and enables the network to focus on the most relevant features.

## Stride

Stride refers to the number of pixels by which the convolutional filter is moved horizontally or vertically across the input image. In other words, it determines the step size of the filter during the convolution operation.

When the stride is set to 1, the filter moves pixel by pixel across the image. This results in output feature maps with the same spatial dimensions as the input. However, a larger stride value, such as 2, moves the filter by two pixels at a time, effectively reducing the size of the output feature map.

Using a larger stride can help reduce computational complexity and downsample the feature maps more aggressively. However, it may also result in some loss of information, as the filter skips certain regions of the input image.




## Basic CNN Architecture

![NN architecture](https://miro.medium.com/v2/resize:fit:1400/format:webp/1*EPpYI-llkbtwHgfprtTJzw.png)




A typical CNN consists of several layers, each serving a specific purpose in the image classification process. The basic architecture includes:

1. **Input Layer**: The input layer receives the raw image data as input. Each pixel of the image is represented by its intensity value (e.g., grayscale image) or a combination of intensity values for each color channel (e.g., RGB image).

2. **Convolutional Layer**: The convolutional layer is responsible for extracting features from the input image using convolutional filters/kernels. These filters capture patterns and detect specific features (such as edges, textures, or shapes) by sliding over the image and performing convolutions.We can use 32 ,64 ,128 filters based on need.

3. **Activation Layer**: An activation layer introduces non-linearity to the network by applying an activation function to the output of the convolutional layer. Common activation functions include ReLU (Rectified Linear Unit), sigmoid, or tanh. Activation functions help in introducing non-linear relationships and allowing the network to learn complex mappings.

4. **Pooling Layer**: The pooling layer reduces the spatial dimensions of the feature maps obtained from the convolutional layer. It helps to retain the most salient features while reducing the computational complexity of the network. Max pooling is a commonly used technique, where the maximum value within a small region (e.g., 2x2) is selected. Pooling helps to achieve translational invariance and reduces the sensitivity of the network to small changes in the position of the features.

5. **Fully Connected Layer**: The fully connected layer takes the output from the previous layers and maps it to the desired number of classes or labels. This layer connects every neuron from the previous layer to every neuron in the current layer, forming a fully connected network. It learns to combine the extracted features from the previous layers and make predictions based on them.

6. **Softmax Layer**: The softmax layer is typically the last layer of the CNN. It applies the softmax activation function to the output of the previous layer, normalizing the values and converting them into probability scores. These probability scores represent the likelihood of the image belonging to each class. The class with the highest probability is considered the predicted class.

By stacking multiple convolutional, activation, pooling, and fully connected layers, CNNs can learn hierarchical representations of the input image. The initial layers capture low-level features like edges and textures, while deeper layers learn more complex and abstract features. This hierarchical representation enables CNNs to classify images accurately.

It's important to note that the architecture of a CNN can vary depending on the specific task or application. More complex architectures, such as VGG, ResNet, or Inception, incorporate additional layers and techniques to improve performance on challenging datasets.

### Importing necessary Package and Loading the Dataset
* we are importing the necessary package and setting the path the dataset directory
* Also we are defining  the image size and batch sie

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
from tensorflow.keras.applications import VGG16, Xception
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization


# Set the path to the dataset
dataset_dir = '/content/'

# Define the image dimensions and batch size
img_height = 256
img_width = 256
batch_size = 32
```

### Creating the generators and spliting the data
* we are using the  tf.keras.utils.image_dataset_from_directory to  creates generators and split it into train and test
* As the number of images are more we can't directly load them as it will consume all the RAM so we use genrators.

```python

seed = 42
# Loading the dataset and spliting into training and testing sets

train_generator = tf.keras.utils.image_dataset_from_directory(
    os.path.join(dataset_dir, 'train'),  # Directory containing the training images
    labels="inferred",                   # Inferred labels from subdirectory names
    label_mode="categorical",            # Use categorical labels (one-hot encoded)
    class_names=None,                    # Automatically infer class names from subdirectory names
    color_mode="rgb",                    # Color mode of the images (RGB)
    batch_size=batch_size,               # Number of images in each batch
    image_size=(img_height, img_width),  # Dimensions to resize the images
    validation_split=0.2,                # Percentage of data to use for validation
    subset="training",                   # Subset of data to retrieve (training set)
    seed=seed                           # Seed for shuffling the data
)


validation_generator = tf.keras.utils.image_dataset_from_directory(
    os.path.join(dataset_dir, 'test'),
    labels="inferred",
    label_mode="categorical",  # Use "categorical" for categorical labels
    class_names=None,
    color_mode="rgb",
    batch_size=batch_size,
    image_size=(img_height, img_width),
    validation_split=0.2,
    subset="validation",
    seed=seed
)
```

# Sample Images

```python
import matplotlib.pyplot as plt

# Function to display sample images from the dataset
def show_sample_images(dataset, class_names, num_images=5):
    plt.figure(figsize=(10, 10))
    for images, labels in dataset.take(1):
        for i in range(num_images):
            ax = plt.subplot(1, num_images, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            plt.title(class_names[np.argmax(labels[i])])
            plt.axis("off")
    plt.show()

# Display sample images from the training set
show_sample_images(train_generator, train_generator.class_names)

# Display sample images from the validation set
show_sample_images(validation_generator, validation_generator.class_names)
```

```python
# Function to count the number of samples in each class
def count_samples(directory):
    files = os.listdir(directory)
    num_samples = len(files)
    return num_samples

# Define directories
train_dogs_dir = os.path.join(dataset_dir, 'train', 'dogs')
train_cats_dir = os.path.join(dataset_dir, 'train', 'cats')

test_dogs_dir = os.path.join(dataset_dir, 'test', 'dogs')
test_cats_dir = os.path.join(dataset_dir, 'test', 'cats')

# Count samples in training and testing sets for dogs and cats
train_dogs_count = count_samples(train_dogs_dir)
train_cats_count = count_samples(train_cats_dir)

test_dogs_count = count_samples(test_dogs_dir)
test_cats_count = count_samples(test_cats_dir)

# Print the counts
print("Training Set:")
print("Dogs:", train_dogs_count)
print("Cats:", train_cats_count)
print()
print("Testing Set:")
print("Dogs:", test_dogs_count)
print("Cats:", test_cats_count)
```

# Normalizing
* The normalization process ensures that the pixel values of the images fall within a consistent range, typically between 0 and 1.
* we are are just converting the images into black and white or grayscale

```python
# Normalizing the data
def process(image,label):
    image = tf.cast(image/255,tf.float32)
    return image,label
train_generator = train_generator.map(process)
validation_generator = validation_generator.map(process)
```

# Creating a CNN architecture

We are using `Xception Model` a highly efficient convolutional neural network architecture that achieves state-of-the-art performance on various computer vision tasks, including image classification and object detection.The Xception model is used as a feature extractor, and its pre-trained weights, obtained from training on the ImageNet dataset, are leveraged to extract meaningful features from input images for the binary classification task at hand.


1. **Xception Model Initialization**:
   - `Xception`: This function initializes the Xception model architecture.
   - `weights='imagenet'`: It specifies that the model should be initialized with pre-trained weights from the ImageNet dataset.
   - `include_top=False`: It excludes the top (classification) layers of the model, which are specific to ImageNet's classes.
   - `input_shape=(256, 256, 3)`: It sets the input shape of the model to (256, 256, 3), representing images with dimensions 256x256 pixels and 3 color channels (RGB).
   - `pooling='avg'`: It sets the global average pooling as the pooling strategy, which reduces the spatial dimensions of the input to a single vector by taking the average of all feature maps.

2. **Freezing Convolutional Base**:
   - `conv_base.trainable = False`: This line freezes the convolutional base (Xception model) to prevent its weights from being updated during training. By freezing the base, only the added layers will be trained.



3. **Building Custom Model**:
   - `Sequential()`: This initializes a sequential model, where layers are added sequentially.
   - `model.add(conv_base)`: This adds the pre-trained Xception model as the first layer of the custom model.
   - `model.add(BatchNormalization())`: This adds a batch normalization layer to normalize the activations of the previous layer.
   - `model.add(Dense(256, activation='relu'))`: This adds a fully connected dense layer with 256 units and ReLU activation function.
   - `model.add(Dropout(0.35))`: This adds a dropout layer to prevent overfitting by randomly setting a fraction of input units to 0 during training.
   - `model.add(BatchNormalization())`: Another batch normalization layer.
   - `model.add(Dense(120, activation='relu'))`: Another fully connected dense layer with 120 units and ReLU activation.
   - `model.add(Dense(2, activation='sigmoid'))`: This adds the output layer with 2 units (assuming binary classification) and sigmoid activation function to output probabilities for two classes (e.g., cat and dog).

This code constructs a neural network architecture for binary classification using transfer learning with the Xception model as the base and additional custom layers on top. The convolutional base (Xception) is kept frozen, while the added layers are trained to adapt to the specific classification task.




```python
# initalizing the Xception model
conv_base = Xception(
    weights='imagenet',
    include_top=False,
    input_shape=(256, 256, 3),
    pooling='avg'
)
conv_base.trainable = False
```

```python
# Creating model
model = Sequential()
model.add(conv_base)
model.add(BatchNormalization())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.35))
model.add(BatchNormalization())
model.add(Dense(120, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(lr=0.0001), loss='binary_crossentropy', metrics=['accuracy'])
```

### Training and Evaluting the model

```python
from tensorflow.keras.callbacks import EarlyStopping

history = model.fit(train_generator, epochs=10, validation_data=validation_generator,callbacks=[EarlyStopping(patience=0)])
```

```python
# Evaluate the model on the validation dataset
validation_loss, validation_accuracy = model.evaluate(validation_generator)

# Print the validation loss and accuracy
print("Validation Loss:", validation_loss)
print("Validation Accuracy:", validation_accuracy)
```
